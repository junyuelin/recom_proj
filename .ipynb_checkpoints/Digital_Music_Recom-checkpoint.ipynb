```python
import pandas as pd
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import SVD
from surprise import accuracy
from surprise.model_selection import GridSearchCV

file_path = 'Digital_Music_5.json'
df = pd.read_json(file_path, lines=True)

reader = Reader(rating_scale=(1, 5))

data = Dataset.load_from_df(df[['reviewerID', 'asin', 'overall']], reader)

trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

param_grid = {'n_epochs': [5, 10, 15], 'lr_all': [0.002, 0.005, 0.01],
              'reg_all': [0.02, 0.04, 0.06]}

gs = GridSearchCV(SVD, param_grid, measures=['rmse'], cv=3)

gs.fit(data)


best_params = gs.best_params['rmse']

print("Best Parameters:", best_params)

best_svd_model = SVD(n_epochs=best_params['n_epochs'], 
                     lr_all=best_params['lr_all'], reg_all=best_params['reg_all'])

best_svd_model.fit(data.build_full_trainset())

predictions = best_svd_model.test(testset)

rmse = accuracy.rmse(predictions)
print("RMSE with Tuned Model:", rmse)
```
